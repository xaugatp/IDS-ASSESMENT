```python
#Now let's select maximum temperature to be the target variable, which will be predicted from rest of the features. 
#weather_x stores the rest of the dataset while weather_y has  temperature column 

weather_df_num = weather_df.loc[:,['Minimum temperature (°C)',
       'Maximum temperature (°C)', 
       'Rainfall (mm)',
       'Evaporation (mm)',
       'Sunshine (hours)',
       '9am Temperature (°C)',
       '9am relative humidity (%)',
       '9am cloud amount (oktas)',
       #'9am wind speed (km/h)',
       '3pm Temperature (°C)', 
       '3pm relative humidity (%)',
       '3pm cloud amount (oktas)',
       #'3pm wind speed (km/h)' 
        ]]
weather_df_num.head()
```

```python
weth = weather_df_num['1/01/2020':'31/12/2020']
weth.head()
```

```python
weather_y = weather_df_num.pop("Maximum temperature (°C)")
weather_x = weather_df_num
```

```python
#now our data is prepared and it is ready to be fed to the model for training it's time to split the dataset into training and testing 
train_X,test_X,train_y,test_y = train_test_split(weather_x,weather_y,test_size=0.2,random_state=4)
```

```python
model=LinearRegression()
model.fit(train_X,train_y)
```

```python

prediction = model.predict(test_X)
```

```python

#calculating error
np.mean(np.absolute(prediction-test_y))
```

```python

print('Variance score: %.2f' % model.score(test_X, test_y))
```

```python
for i in range(len(prediction)):
  prediction[i]=round(prediction[i],2)
pd.DataFrame({'Actual':test_y,'Prediction':prediction,'diff':(test_y-prediction)})
```

```python
from sklearn.metrics import r2_score
print("Mean absolute error: %.2f" % np.mean(np.absolute(prediction - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((prediction - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y,prediction ) )
```
